```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as K
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
from tensorflow.keras.applications.inception_v3 import InceptionV3
%matplotlib inline
```

```python
import os
import zipfile
from google.colab import drive
drive.mount("/content/drive")
```

```python
IMAGE_SIZE = [224,224]
```

```python
#train_path = '/content/drive/MyDrive/dataset cats & Dogs/training_set'
#test_path = '/content/drive/MyDrive/dataset cats & Dogs/test_set'

#train_path = '/content/drive/MyDrive/sample dataset/train'
#test_path = '/content/drive/MyDrive/sample dataset/test'

train_path = '/content/drive/MyDrive/PetImages/train images'
test_path = '/content/drive/MyDrive/PetImages/test images'
```

```python

inception = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
for layer in inception.layers:
    layer.trainable = False
```

```python
# useful for getting number of output classes
folders = glob('/content/drive/MyDrive/PetImages/train images/*')
```

```python
folders
```

```python
len(folders)
```

```python
# our layers - you can add more if you want
x = Flatten()(inception.output)
```

```python
prediction = Dense(len(folders), activation='softmax')(x)

# create a model object
model = Model(inputs=inception.input, outputs=prediction)
```

```python
model.summary()
```

```python
# tell the model what cost and optimization method to use
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
# Use the Image Data Generator to import the images from the dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# Make sure you provide the same target size as initialied for the image size
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/PetImages/train images',
                                                 target_size = (224, 224),
                                                 batch_size = 16,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/PetImages/test images',
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
print(len(test_set))
len(training_set)
```

```python
# fit the model
# Run the cell. It will take some time to execute
r = model.fit_generator(
  training_set,
  validation_data=test_set,
  epochs=1,
  steps_per_epoch=len(training_set),
  validation_steps = len(test_set)
)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot the loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```

```python
y_pred = model.predict(test_set)
y_pred
```
